```python
import pandas as pd
import numpy as np
df = pd.read_csv("/home/student_user/Downloads/Titanic-Dataset.csv")
print(df.info())
```

```python
df.drop("Name",axis=1,inplace=True)
df.drop("Ticket",axis=1,inplace=True)
df.drop("PassengerId",axis=1,inplace=True)
df.drop("Cabin",axis=1,inplace=True)
df.drop("Embarked",axis=1,inplace=True)
```

```python
print(df.info())
```

```python
print(df.isnull().sum())
```

```python
updated_df = df
updated_df['Age']=updated_df['Age'].fillna(updated_df['Age'].mean())
updated_df.info()
```

```python
y1 = updated_df["Survived"]
```

```python
updated_df.info()
```

```python
updated_df.drop("Survived",axis=1,inplace=True)
```

```python
updated_df["Label"] = pd.cut(x=updated_df['Age'],bins=[0,13,17,63,100], labels=["Baby","Child","Adult","Elder"])
```
