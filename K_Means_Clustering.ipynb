# Task-1:



```python
from sklearn.datasets import load_iris
iris = load_iris()
x = iris.data
y = iris.target

feature_names = iris.feature_names
target_name = iris.target_names

print("Feature names:", feature_names)
print("target names:", target_name)
print("First 10 rows of x:", x[ : 10])
```

```python
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import sklearn.metrics as sm
```

```python
x=pd.DataFrame(iris.data, columns=["SL", "SW", "PL", "PW"])
y=pd.DataFrame(iris.target, columns=["Target"])
```

```python
plt.figure(figsize=(12,3))
colors =np.array(["red", "green", "blue"])
iris_target_legends =np.array(iris.target_names)

red_patch = mpatches.Patch(color="red", label="setosa")
green_patch = mpatches.Patch(color="green", label="versicolor")
blue_patch = mpatches.Patch(color="blue", label="virginica")

plt.subplot(1,2,1)
plt.scatter(x["SL"], x["SW"], c=colors[y["Target"]])
plt.title("Sepal Length vs Petal Length visualization")
plt.legend(handles=[red_patch, green_patch, blue_patch])

plt.subplot(1,2,2)
plt.scatter(x["PL"], x["PW"], c=colors[y["Target"]])
plt.legend(handles=[red_patch, green_patch, blue_patch])
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2,random_state=42)
```

```python
model= KMeans(n_clusters=3)
model.fit(x_train, y_train)
```

```python
y_pred=model.predict(x_test)
print(y_pred, y_test)
```

# Task-2:

```python
accuracy = sm.accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
